Балансировка классов автоматически с присвоением весов привела к худшим результатам, чем ручная. Поэтому убираем дисбаланс вручную в следующих моделях.

Было решено оптимизировать метрику F1 macro, так как нам важна точность предсказания каждого класса.

```python
from sklearn.preprocessing import LabelEncoder

# Балансировка классов
def balance_classes(df, target_col, samples_per_class=2000):
    balanced_df = pd.DataFrame()
    for class_name in df[target_col].unique():
        class_df = df[df[target_col] == class_name]
        if len(class_df) > samples_per_class:
            class_df = class_df.sample(samples_per_class, random_state=42)
        balanced_df = pd.concat([balanced_df, class_df])
    return balanced_df


balanced_df = balance_classes(df, 'genre')
balanced_df = balanced_df.drop(columns=["artist_name", "track_name", "track_id", 'popularity', "year"])
le = LabelEncoder()
balanced_df["genre"] = le.fit_transform(balanced_df["genre"])

X = balanced_df.drop("genre", axis=1)
y = balanced_df["genre"]
```

```python
from sklearn.model_selection import train_test_split
import optuna
from optuna.samplers import TPESampler
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, f1_score
import pandas as pd
import matplotlib.pyplot as plt

# Разбиваем данные
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

def objective(trial):
    params = {
        'objective': 'multi:softprob',
        'num_class': len(le.classes_),
        'eval_metric': 'mlogloss',
        'verbosity': 0,
        'tree_method': 'gpu_hist',
        'gpu_id': 0,
        'n_estimators': trial.suggest_int('n_estimators', 100, 500),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'gamma': trial.suggest_float('gamma', 0, 1),
        'subsample': trial.suggest_float('subsample', 0.6, 1),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1),
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 1),
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 1),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 0.8, 1.2),
    }

    model = XGBClassifier(**params, random_state=42)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return f1_score(y_test, preds, average='macro')

# Настройка Optuna
study = optuna.create_study(
    direction='maximize',
    sampler=TPESampler(seed=42),
    pruner=optuna.pruners.HyperbandPruner()
)

study.optimize(objective, n_trials=50, show_progress_bar=True)

# Лучшие параметры и метрики
print("\nЛучший Macro F1:", study.best_value)
print("\nЛучште парам:", study.best_params)

# Финальная модель
best_model = XGBClassifier(
    **study.best_params,
    objective='multi:softprob',
    num_class=len(le.classes_),
    tree_method='gpu_hist',
    gpu_id=0,
    random_state=42
)
best_model.fit(X_train, y_train)

# Полная оценка
y_pred = best_model.predict(X_test)
print("\nDetailed Classification Report:")
print("="*60)
report = classification_report(y_test, y_pred, target_names=le.classes_, digits=4, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print(report_df.to_markdown())
print("="*60)

# Визуализация метрик по классам
plt.figure(figsize=(10, 6))
metrics = ['precision', 'recall', 'f1-score']
for metric in metrics:
    plt.plot(le.classes_, [report[cls][metric] for cls in le.classes_],
             marker='o', label=metric)
plt.title('Metrics per Class')
plt.xlabel('Class')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# Важность признаков
feat_imp = pd.DataFrame({
    'feature': best_model.feature_name_,
    'importance': best_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\nВажность признаков:")
print(feat_imp.to_markdown())


plt.figure(figsize=(12, 8))
feat_imp.head.sort_values('importance').plot.barh(x='Признак', y='Важность')
plt.show()
```

```python
from lightgbm import LGBMClassifier

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

def objective(trial):
    params = {
        'objective': 'multiclass',
        'num_class': len(np.unique(y)),
        'metric': 'multi_logloss',
        'verbosity': -1,
        'boosting_type': 'gbdt',
        'device': 'gpu',
        'gpu_platform_id': 0,
        'gpu_device_id': 0,
        'n_estimators': trial.suggest_int('n_estimators', 100, 500),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'num_leaves': trial.suggest_int('num_leaves', 20, 100),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'min_child_samples': trial.suggest_int('min_child_samples', 10, 100),
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 1),
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 1),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1),
        'subsample': trial.suggest_float('subsample', 0.6, 1),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 0.8, 1.2),
    }

    model = LGBMClassifier(**params, random_state=42)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return f1_score(y_test, preds, average='macro')

# Настройка Optuna
study = optuna.create_study(
    direction='maximize',
    sampler=TPESampler(seed=42),
    pruner=optuna.pruners.HyperbandPruner()
)

study.optimize(objective, n_trials=50, show_progress_bar=True)

# Лучшие параметры и метрики
print("\nBest Macro F1:", study.best_value)
print("Best params:", study.best_params)

# Финальная модель
best_model = LGBMClassifier(
    **study.best_params,
    objective='multiclass',
    num_class=len(np.unique(y)),
    device='gpu',
    random_state=42
)
best_model.fit(X_train, y_train)

# Полная оценка
y_pred = best_model.predict(X_test)
print("\nDetailed Classification Report:")
print("="*60)
report = classification_report(y_test, y_pred, target_names=le.classes_, digits=4, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print(report_df.to_markdown())
print("="*60)

# Визуализация метрик по классам
plt.figure(figsize=(10, 6))
metrics = ['precision', 'recall', 'f1-score']
for metric in metrics:
    plt.plot(le.classes_, [report[cls][metric] for cls in le.classes_],
             marker='o', label=metric)
plt.title('Metrics per Class')
plt.xlabel('Class')
plt.ylabel('Score')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# Важность признаков
feat_imp = pd.DataFrame({
    'feature': best_model.feature_name_,
    'importance': best_model.feature_importances_
}).sort_values('importance', ascending=False)

print("\nВажность признаков:")
print(feat_imp.to_markdown())

# Визуализация важности признаков
plt.figure(figsize=(12, 8))
feat_imp.head.sort_values('importance').plot.barh(x='Признак', y='Важность')
plt.show()
```

Для улучшения предсказания жанра в дальнейшем можно будет использовать более сильные архитектуры на более мощном gpu, но обучение займет слишком много времени и ресурсов. Если не обязательно сохранять все жанры, то возможно стоит объединить малочисленные и похожие жанры в одну группу, удалить малоэффективные признаки, сгенерировать новые, добавить кросс-валидацию и попробовать сделать ансамбль моделей.
